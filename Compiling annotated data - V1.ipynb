```python
labels_dic = {14: "Source",
             15: "Action from source",
             16: "Reported Statement"}
new_dic = {'id': [], 'Source': [], "Action from source": [], "Reported Statement" : [], 'Original text' : []}
f = lambda x: x.dropna()
```

```python
def read_annotations(text, labels, fill_df):
    for i, dic in enumerate(labels):
        test = [i, np.nan, np.nan, np.nan, text]
        j = list(empty_df.columns).index(labels_dic[dic['label']])
        test[j] = text[dic['start_offset']: dic['end_offset']]
        fill_df = fill_df.append(dict(zip(empty_df.columns,test)), ignore_index=True)
    fill_df['group'] = fill_df.id.apply(lambda x: 0)
    #fill_df = fill_df.groupby(fill_df.group).agg({'Source': f,
    #                                          'Action from source': f,
    #                                          'Reported Statement': f,
    #                                          'Original text': lambda x: x.drop_duplicates()}).reset_index()
    return fill_df
```

```python
empty_df = pd.DataFrame(new_dic)
i=1
text, labels = df[['text', 'annotations']].loc[i]
test = read_annotations(text, labels, empty_df)
test

i = 0
g = 0
while i <= test.shape[0]:
    t = test[test.group==g][0:0+3]
    m = t[['Source', 'Action from source', 'Reported Statement']].isna().sum().min()
    if m<2:
        test.group[i:i+2] = [g, g]
        test.group[i+2:] = [g+1]*test.group[i+2:].shape[0]
        i=i+2
    elif m>=2:
        test.group[i:i+3] = [g, g, g]
        test.group[i+3:] = [g+1]*test.group[i+3:].shape[0]
        i=i+3
    g=g+1
```
